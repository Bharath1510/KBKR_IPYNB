# 2. Write a program to print twin primes less than 1000.

```python
# import sqrt func. from math module
from math import sqrt

def primeNum(N):
    n = int(sqrt(N))#finding int(sqrt) and initialising it to n
    #iterating over range return if prime or not
    for i in range(2,n+1):
        if (N%i) == 0:
            return False
    else:#return True if prime
        return True
#for loop to print twin prime less than 1000    
for n in range(3,1001,2):
    #checking two consecutive odd numbers are prime or not by using primeNum func.
    if primeNum(n) and primeNum(n+2):
        #printing twin primes as a tuple
        print("({},{})".format(n,n+2),end=" ")
   
    
```

# 3. Write a program to find out the prime factors of a number.

```python
# import sqrt func. from math module
from math import sqrt

def primeFac(N):
    lst = []#initializing empty list to store prime factors
    #divide by 2 till N becomes odd
    while N%2==0:
        N=int(N/2)
        lst.append(2)# append factor to the list
    for i in range(3,N+1,2):
        while N%i==0:#checking for prime factors and appending
            lst.append(i)
            N=int(N/i)
    
    return lst  # to return list of prime factors
N = int(input("Enter the number for Prime factors: "))
primeFac(N)
```

```python
def fact(N):
    #initializing variables to find factorial
    i=1
    f=1
    if N==1 or N==2:# factorial of 1 & 2 is the number itself
        return N
    while i<=N:# iterate from 1 to N to find factorial
        f = f*i
        i+=1
    return f

#using fact in permutations formula
def permutations(n,r):
    p = fact(n)/fact(n-r)
    return p
#using fact in combinations formula
def combinations(n,r):
    c = fact(n)/(fact(r)*fact(n-r))
    return c

n = int(input("Enter n: "))
r = int(input("Enter r: "))
print("permutations: ",permutations(n,r))
print("combinations: ",combinations(n,r))
```

# 5. Write a function that converts decimal number to binary.

```python
def dec_bin(N):
    """
    Using predefined functions
    binary = bin(N)[2:]
    return int(binary)
    """
    lst = []#initializing empty list to store binary
    while N>0:
        i = N%2
        N = N//2
        lst.append(str(i))#doing LCM with 2 and storing binary remainders in lst
    binary = "".join(reversed(lst))#converting lst to string
    print(binary)
    
N = int(input("Enter the number to be converted to binary: "))
dec_bin(N)
```

```python
def cubeSum(N):
    cube = 0#intializing cube to 0
    while N>0:
        #accessing digits and calculating sum of cubes of digits
        digit = N%10
        cube += digit**3
        N = N//10
    return cube
#to find cubeSum is Armstrong or not
def isArmstrong(N):
    # to check condition of Armstrong number
    if N == cubeSum(N):
        return True
    else:
        return False
#to find armstrong numbers in given range using isArmstrong function
def printArmstrong(low,upper):
    lst = []
    for num in range(low,upper+1):
        if isArmstrong(num):
            lst.append(num)
    return lst
            
N = int(input("enter the number: "))
print("Sum of cubes of digits: ",cubeSum(N))
print("is it a Armstrong number: ",isArmstrong(N))

printArmstrong(1,1000)
        
```

# 7. Write a function prodDigits() that inputs a number and returns the product of digits of that number.

```python
def prodDigits(num):
    prod = 1#initializing prod to 1
    while num>0:
        # finding individual digits and multiplying digits with other digits without repitition
        i = num % 10
        prod *= i
        num //=10
    return prod# returning product of digits

num = int(input("enter the number: "))
prodDigits(num)
```

```python
#similar implementation as above proddigits
def prodDigits(num):
    prod = 1
    while num>0:
        i = num % 10
        prod *= i
        num //=10
    return prod

def MDR(num):
    #iterating over product of digits till single digit is reached
    while len(str(num))!=1:
        num = prodDigits(num)
    return num

def MPersistence(num):
    lst = []#initilizing empty list to store prodDigits
    while len(str(num))!=1:
        lst.append(num)#appending prodDigits to the list
        num = prodDigits(num)
    return len(lst)#return length of list to find multiplicative digital root

num = int(input("enter the number: "))
print("Multiplicative digital root: ",MDR(num))
print("Multiplicative persistence: ",MPersistence(num))
```
